```python
import csv

CUTOFF = 0.003

with open("Dump_RS485_Uhrzeit.txt", "r") as packet_capture:
    packetreader = csv.reader(packet_capture, delimiter=',', quotechar='"')
    data = list(packetreader)[1:]

data = list(map(lambda l: (float(l[0]), int(l[1], 16)), data))

byte_buffer = []
packets = []

for idx in range(len(data)-1):
    byte_buffer.append(data[idx][1])
    diff = data[idx + 1][0] - data[idx][0]
    
    if diff > CUTOFF:
        packets.append(byte_buffer)
        byte_buffer = []

packets.append(byte_buffer)
```

```python
from functools import reduce
import operator

print("#packets: " + str(len(packets[1:])) + "\n\n")

for p in packets[1:]:
    hexpacket = list(map(hex, p))

    # parse Sender
    sender = "";
    match hexpacket[0]:
        case "0x55":
            sender = 'B'
        case "0x1":
            sender = 'A'
        case _:
            sender = hexpacket[0]
        
    print("Sender: " +  sender)
    print("length: " + str(len(hexpacket)))
    
    print(hexpacket[1:])
    print(list(map(chr, p[1:])))
    print("------------------------------------")
```
